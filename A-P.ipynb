```python
import random
import time
import tkinter as tk
```

```python

M = 9

def puzzle(a):
    for i in range(M):
        for j in range(M):
            print(a[i][j], end=" ")
        print()

def solve(grid, row, col, num):
    for x in range(M):
        if grid[row][x] == num:
            return False

    for x in range(M):
        if grid[x][col] == num:
            return False

    startRow = row - row % 3
    startCol = col - col % 3
    for i in range(3):
        for j in range(3):
            if grid[i + startRow][j + startCol] == num:
                return False
    return True

def Sudoku(grid, row, col):
    if (row == M - 1 and col == M):
        return True
    if col == M:
        row += 1
        col = 0
    if grid[row][col] > 0:
        return Sudoku(grid, row, col + 1)
    for num in range(1, M + 1):
        if solve(grid, row, col, num):
            grid[row][col] = num
            if Sudoku(grid, row, col + 1):
                return True
            grid[row][col] = 0
    return False

def get_user_input():
    num1 = int(entry1.get())
    num2 = int(entry2.get())
    return num1, num2

def generate_empty_sudoku():
    grid = [[0] * M for _ in range(M)]
    return grid

def fill_random_sudoku(grid):
    for i in range(M):
        for j in range(M):
            if grid[i][j] == 0:
                num = random.randint(1, M)
                if solve(grid, i, j, num):
                    grid[i][j] = num
                else:
                    grid[i][j] = 0

def solve_sudoku():
    grid = generate_empty_sudoku()
    num1, num2 = get_user_input()
    grid[0][0] = num1
    grid[1][1] = num2
    fill_random_sudoku(grid)
    if Sudoku(grid, 0, 0):
        puzzle(grid)
    else:
        print("Solution does not exist:(")
    end_time = time.time()
    execution_time = end_time - start_time
    print("زمان اجرا: ", execution_time, "ثانیه")
    print(grid)
    label_result["text"] = "زمان اجرا: {} ثانیه".format(execution_time)
    display_grid(grid)  # نمایش ماتریس در پنجره

def display_grid(grid):
    for i in range(M):
        for j in range(M):
            label_grid[i][j]["text"] = str(grid[i][j])

# تنظیمات رابط کاربری
window = tk.Tk()
window.title("Sudoku Solver")

label1 = tk.Label(window, text="Enter the first number (1-9):")
label1.grid(row=0, column=0)

entry1 = tk.Entry(window)
entry1.grid(row=0, column=1)

label2 = tk.Label(window, text="Enter the second number (1-9):")
label2.grid(row=1, column=0)

entry2 = tk.Entry(window)
entry2.grid(row=1, column=1)

solve_button = tk.Button(window, text="Solve", command=solve_sudoku)
solve_button.grid(row=2, columnspan=2)

label_result = tk.Label(window, text="")
label_result.grid(row=3, columnspan=2)

label_grid = []
for i in range(M):
    row = []
    for j in range(M):
        label = tk.Label(window, width=2, relief="solid")
        label.grid(row=i+4, column=j)
        row.append(label)
    label_grid.append(row)

start_time = time.time()

window.mainloop()
```
